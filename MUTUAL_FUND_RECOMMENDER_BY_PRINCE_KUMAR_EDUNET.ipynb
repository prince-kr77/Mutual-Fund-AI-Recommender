```python
df = pd.read_csv(io.BytesIO(uploaded['MUTUAL_FUNDS_DATASET.csv']))
```

```python
print("‚úÖ File loaded successfully!")
print(df.head())
```

```python
features = ['AUM(In Rs. Cr)', 'Expenseratio (%)', 'Return (%)1 Yr', 'NAV','Alpha', 'Avg. Maturity(In Yrs)',
           'Yield To Maturity (%)', 'Sharpe', 'Sortino', 'Beta', 'Standard Deviation', 'Fund Age (Years)']
df.dropna(subset=features, inplace=True)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df[features])
```

```python
knn = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn.fit(X)
```

```python
print("\nüßë Please answer the following questions:")

user_input = pd.DataFrame({
    'AUM(In Rs. Cr)': [float(input("üëâ Minimum Asset Under Management (AUM) you prefer? (e.g., 1000): "))],
    'Expenseratio (%)': [float(input("üëâ Max Expense Ratio (%) (e.g., 1.5): "))],
    'Return (%)1 Yr': [float(input("üëâ Expected 1-Year Return (%) (e.g., 12): "))],
    'NAV': [float(input("üëâ Desired NAV (e.g., 50): "))],
    'Alpha': [float(input("üëâ Expected Alpha (e.g., 2.0): "))],
    'Avg. Maturity(In Yrs)': [float(input("üëâ Preferred Average Maturity (years) (e.g., 5): "))],
    'Yield To Maturity (%)': [float(input("üëâ Yield to Maturity (%) (e.g., 7): "))],
    'Sharpe': [float(input("üëâ Desired Sharpe Ratio? (e.g., 1.2): "))],
    'Sortino': [float(input("üëâ Desired Sortino Ratio (e.g., 1.5): "))],
    'Beta': [float(input("üëâ Preferred Beta (lower means lower volatility (e.g., 0.9): "))],
    'Standard Deviation': [float(input("üëâTolerance for Standard Deviation (risk) (e.g., 10): "))],
    'Fund Age (Years)': [float(input("üëâ Minimum Fund Age (e.g., 3): "))]
})
```

```python
user_input_scaled = scaler.transform(user_input)
```

```python
distances, indices = knn.kneighbors(user_input_scaled)
```

```python
print("\nüéØ Top 5 Recommended Mutual Funds For You:")
recommended = df.iloc[indices[0]]
print(recommended[['Funds', 'Classification', 'Fund Type', 'AUM(In Rs. Cr)', 'Return (%)1 Yr',
                   'Sharpe', 'Alpha', 'Fund Age (Years)']])
```
